Смотрим только США
---

```python
mask1 = df_cut['Географическое положение'] == 'United States'
```

```python
df_cut_USA = df_cut[mask1]
```

```python
df_cut_USA.head()
```

Получение датасета корреляций
---

```python
corr_matrix_usa = df_cut_USA.corr().round(5)
```

```python
s = corr_matrix_usa.unstack()
so = s.sort_values()
```

```python
sq = pd.DataFrame()
sq['colocations'] = so.index.to_list()
sq['values'] = so.to_list()
```

```python
max_corr_USA = {}
```

```python
for i, a in zip(sq['colocations'], sq['values']):
    if (a >= 0.8 and a < 1) or (a <= -0.8 and a > -1):
      max_corr_USA[i] = a
      print(i, a)
```

```python
max_USA = pd.DataFrame.from_dict(max_corr_USA, orient='index')
```

```python
max_USA.to_excel('max_USA.xlsx')
```
